**Laporan klasifikasi & ROC-AUC multi-kelas.** `classification_report` merangkum precision/recall/F1; `predict_proba` + `label_binarize` memungkinkan `roc_auc_score` dengan strategi one-vs-one (`average="macro"`).

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, RocCurveDisplay, f1_score

X, y = load_breast_cancer(return_X_y=True)  # klasifikasi biner
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
clf = LogisticRegression(max_iter=5000)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
print("F1 macro", f1_score(y_test, y_pred, average="macro"))
RocCurveDisplay.from_estimator(clf, X_test, y_test)
```
