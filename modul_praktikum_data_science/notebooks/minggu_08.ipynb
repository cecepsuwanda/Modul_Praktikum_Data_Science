**Bandingkan tiga klasifikator.** Iris di-split stratifikasi; loop melatih LogReg, KNN ($k=5$), dan pohon keputusan lalu mencetak akurasi test---bandingkan dengan modul PDF minggu 8 (confusion matrix lanjutan bisa ditambahkan).

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

log_reg = LogisticRegression(max_iter=200)
knn = KNeighborsClassifier(n_neighbors=5)
tree = DecisionTreeClassifier(max_depth=3, random_state=42)

for name, m in [("logreg", log_reg), ("knn", knn), ("tree", tree)]:
    m.fit(X_train, y_train)
    print(name, m.score(X_test, y_test))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

pred = log_reg.predict(X_test)
cm = confusion_matrix(y_test, pred)
ConfusionMatrixDisplay(cm).plot()
plt.title("Confusion matrix (LogReg)")
plt.show()
```
