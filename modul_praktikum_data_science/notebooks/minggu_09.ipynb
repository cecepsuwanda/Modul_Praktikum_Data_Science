# Minggu 9 — Random Forest vs baseline

**Sub-CPMK:** **P8**


**Random Forest vs baseline.** `DummyClassifier(strategy="most_frequent")` memprediksi kelas modus train; Random Forest ensemble pohon seharusnya mengungguli baseline pada Iris. `n_estimators` mengatur jumlah pohon.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.ensemble import RandomForestClassifier

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
base = DummyClassifier(strategy="most_frequent")
base.fit(X_train, y_train)

rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)
print("baseline", base.score(X_test, y_test))
print("rf      ", rf.score(X_test, y_test))
```

**Importansi fitur.** Isi sel ini: `pd.Series(rf.feature_importances_).sort_values(ascending=False)` — jelaskan bahwa angka itu heuristik, bukan bukti kausal.

```python
import pandas as pd

imp = pd.Series(rf.feature_importances_, index=load_iris().feature_names)
print(imp.sort_values(ascending=False))
```
