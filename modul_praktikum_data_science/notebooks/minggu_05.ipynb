```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
df = sns.load_dataset("penguins").dropna()
df.head()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
sns.histplot(df["bill_length_mm"], kde=True, ax=ax[0])
sns.scatterplot(data=df, x="bill_length_mm", y="bill_depth_mm", hue="species", ax=ax[1])
plt.tight_layout()
plt.show()
```
