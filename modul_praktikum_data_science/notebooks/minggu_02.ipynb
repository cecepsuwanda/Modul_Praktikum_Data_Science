**Memuat dan mengintip data.** `sns.load_dataset` mengembalikan DataFrame Pandas untuk dataset bawaan Seaborn. `df.head(10)` menampilkan 10 baris pertama Titanic untuk memeriksa nilai dan nama kolom; `print` di atasnya memberi konteks teks di antara output tabel.

```python
import pandas as pd  # alias umum untuk DataFrame dan I/O
import seaborn as sns  # dataset contoh + visualisasi

df_tips = sns.load_dataset("tips")  # DataFrame restoran (ujung proporsi / regresi)
df_titanic = sns.load_dataset("titanic")  # DataFrame penumpang (klasifikasi umum)

# Aktifkan baris di bawah untuk ringkas dimensi dan tipe tips:
print("tips", df_tips.shape)
print(df_tips.dtypes.head())
print("10 data dari tips dataset : ")
df_tips.head(10)  # tampilkan tips jika perlu
# df_titanic.head(10)  # output terakhir: tampilan HTML 10 baris pertama Titanic
```

**Ringkasan struktur Titanic.** `DataFrame.info()` mencetak jumlah non-null per kolom, tipe (`dtype`), dan memori—berguna mendeteksi missing dan kolom bertipe `object` yang seharusnya numerik. Baris `describe` yang dikomentari bisa diaktifkan untuk statistik ringkas semua tipe.

```python
print(df_titanic.info())  # ringkasan kolom, non-null count, dtype
# df_titanic.describe(include="all").T.head(14)  # statistik deskriptif (aktifkan untuk eksplorasi)
```

**Sumber terbuka lewat OpenML.** `fetch_openml(..., as_frame=True)` mengunduh dataset Iris sebagai struktur sklearn; atribut `.frame` adalah DataFrame pandas. Perlu koneksi internet pada unduhan pertama; versi kolom bisa sedikit berbeda dari Iris sklearn klasik.

```python
from sklearn.datasets import fetch_openml  # akses dataset publik OpenML

iris_oml = fetch_openml(name="iris", version=1, as_frame=True)  # as_frame=True -> DataFrame di .frame
df_iris = iris_oml.frame  # fitur + target dalam satu tabel
print(df_iris.shape, df_iris.columns.tolist())  # dimensi + nama kolom
df_iris.head()  # lima baris pertama
```

**Ringkasan perbandingan DataFrame.** Fungsi `ringkas` mengembalikan kamus dengan jumlah baris, kolom, dan total sel hilang (`isna`). Membandingkan beberapa sumber membantu Anda memilih dataset untuk tugas berikutnya.

```python
def ringkas(nama, frame):  # nama: label; frame: DataFrame
    return {
        "nama": nama,
        "baris": len(frame),  # jumlah baris
        "kolom": frame.shape[1],  # jumlah kolom
        "missing_total": int(frame.isna().sum().sum()),  # total sel NaN
    }


print(ringkas("tips", df_tips))
print(ringkas("titanic", df_titanic))
print(ringkas("iris_oml", df_iris))
```

**CSV dari URL.** `pd.read_csv(url)` mem-parse CSV langsung dari web; pastikan lisensi dan kestabilan URL. `shape` memberi dimensi; `head(3)` memverifikasi kolom (mis. nama kapital) dibanding variasi Titanic dari Seaborn.

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"  # CSV publik (periksa lisensi)
df_url = pd.read_csv(url)  # unduh + parse menjadi DataFrame
print(df_url.shape)  # (baris, kolom) — skema bisa beda dari sns.load_dataset("titanic")
df_url.head(3)  # tiga baris awal
```
