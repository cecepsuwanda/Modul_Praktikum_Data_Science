```python
import pandas as pd
import seaborn as sns

df_tips = sns.load_dataset("tips")
df_titanic = sns.load_dataset("titanic")

print("tips", df_tips.shape, df_tips.dtypes.head())
print("titanic", df_titanic.shape)
df_titanic.head(10)
```

```python
print(df_titanic.info())
df_titanic.describe(include="all").T.head(14)
```

```python
from sklearn.datasets import fetch_openml

iris_oml = fetch_openml(name="iris", version=1, as_frame=True)
df_iris = iris_oml.frame
print(df_iris.shape, df_iris.columns.tolist())
df_iris.head()
```

```python
def ringkas(nama, frame):
    return {
        "nama": nama,
        "baris": len(frame),
        "kolom": frame.shape[1],
        "missing_total": int(frame.isna().sum().sum()),
    }


print(ringkas("tips", df_tips))
print(ringkas("titanic", df_titanic))
print(ringkas("iris_oml", df_iris))
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df_url = pd.read_csv(url)
print(df_url.shape)
df_url.head(3)
```
