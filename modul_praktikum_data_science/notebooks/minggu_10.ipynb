# Minggu 10 — K-Means & silhouette

**Sub-CPMK:** **P8** (unsupervised)


**K-Means & siluet.** Label asli Iris diabaikan (`_`); `fit_predict` memberi klaster 0..K-1; `silhouette_score` mengukur seberapa terpisah klaster (lebih tinggi biasanya lebih baik, tergantung bentuk data).

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

X, _ = load_iris(return_X_y=True)  # unsupervised: target asli tidak dipakai
Xs = StandardScaler().fit_transform(X)

km = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = km.fit_predict(Xs)
print("inertia", km.inertia_)
print("silhouette", silhouette_score(Xs, labels))
```

**Sweep `n_clusters`.** Latihan: ulang KMeans untuk k=2..6, kumpulkan inertia atau siluet; ganti `pass` dengan loop atau plot.

```python
from sklearn.cluster import KMeans

scores = []
for k in range(2, 8):
    kmk = KMeans(n_clusters=k, random_state=42, n_init=10)
    lab = kmk.fit_predict(Xs)
    scores.append((k, kmk.inertia_, silhouette_score(Xs, lab)))

for k, inert, sil in scores:
    print(k, "inertia", inert, "silhouette", sil)
```

```python
import pandas as pd

dfp = pd.DataFrame(Xs, columns=load_iris().feature_names)
dfp["cluster"] = labels
print(dfp.groupby("cluster").mean())
```
