# Minggu 10 — K-Means & silhouette

**Sub-CPMK:** **P8** (unsupervised)


**K-Means & siluet.** Label asli Iris diabaikan (`_`); `fit_predict` memberi klaster 0..K-1; `silhouette_score` mengukur seberapa terpisah klaster (lebih tinggi biasanya lebih baik, tergantung bentuk data).

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X, _ = load_iris(return_X_y=True)  # unsupervised: target asli tidak dipakai
km = KMeans(n_clusters=3, random_state=int(42), n_init=10)  # n_init: berapa inisialisasi centroid (versi sklearn Anda)
labels = km.fit_predict(X)  # label klaster per sampel
silhouette_score(X, labels)  # skor rata-rata siluet (output terakhir sel)
```
